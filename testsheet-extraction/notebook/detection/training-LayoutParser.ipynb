```python
script = """
import layoutparser as lp
import cv2
import os
import time
lmap = {0: "heading", 1: "question", 2: "subquestion", 3: "choice", 4:"image", 5:"table", 6:"blank", 7:"auxillary_text"}

model = lp.models.Detectron2LayoutModel('/content/cfg.yaml', 
                                 extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.71],
                                 label_map=lmap)

paths = [x.path for x in os.scandir("/content/small_test") if x.path.split(".")[-1] == "png"]
print(len(paths))


sumt = 0
for path in paths:
    print(f"predic---> {path}")
    image = cv2.imread(path)[..., ::-1]
    st = time.time()
    layout = model.detect(image)
    sumt += time.time() - st
    canvas = lp.draw_box(image, layout, box_width=3)
    canvas.save(os.path.join("results", path.split("/")[-1]))

print("total infer time = ", sumt / len(paths))



if __name__ == "__main__": 
    pass

"""

with open("s.py" , "w") as f: 
    f.write(script)
```
