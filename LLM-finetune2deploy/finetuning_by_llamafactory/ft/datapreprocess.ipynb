````python
import pandas as pd
from datasets import Dataset
from trl import SFTTrainer, SFTConfig
import json, re, random

def generate_conversation(examples, rubric=structured_text_rubric2):
    instruction = f"""
You are an Ielts speaking reviewer on Lexical Resource.
Given a question and  the rubric, you evaluate the student's answers and give them a score.
======================
The Lexical Resource rubric includes:
```
{rubric}

```
"""
    conversations = []
    for example in examples:
        resonning = example["output"].split("</think>")[0].replace("<think>", "").strip()
        score = example["output"].split("</think>")[1].replace("Therefore,", "").strip()
        conversations.append({
            "instruction": instruction,
            "input" : example["input"].replace("Evaluate", "Score").replace("evaluated", "scored"),
            "output" : score,
        })
        if random.random() > 0.5:
            conversations.append({
                        "instruction": instruction,
                        "input" : example["input"].replace("Evaluate", "Analysis"),
                        "output" : resonning,
                    })
    return conversations
def read_jsonfile(file_path=""):
    all_samples = []
    # Open and read the file line by line
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            data = json.loads(line)  # Parse the JSON object from each line
            # print(data)              # Do something with the JSON object
            all_samples += [data]
    return all_samples
````

````python
def save_to_jsonl(data, filename):
    try:
        with open(filename, 'w', encoding='utf-8') as file:
            for item in data:
                json.dump(item, file)
                file.write('\n')  # Write each dictionary as a separate line
        print(f"Data saved successfully into {filename}")
    except Exception as e:
        print(f"An error occurred: {e}")

save_to_jsonl(dataX, "/home/ducanh/nvidia-llm-pipeline/llamafactory/data/gra.json")
````
